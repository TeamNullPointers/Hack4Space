# California EPA State Agency CO2
https://data.ca.gov/dataset/calepa-state-agency-co2 

```python
cali_co2 = pd.read_csv("./data/state-agency-co2e-data-2010-2014.csv")
```

```python
cali_co2.shape
```

```python
cali_co2.describe(include="all")
```

```python
cali_co2.head()
```

Can I organize by year (bins) and add up CO2 per year?

```python
co2_years = cali_co2['Emission Year'].unique()
co2_years = np.sort(co2_years)

print(co2_years)
```

```python
co2_e = {}
for year in co2_years:
    co2_e[year] = 0
```

```python
for key, row in cali_co2.iterrows():
    year = row["Emission Year"]
    co2e = row["CO2e"]
    if co2e > 0:
        co2_e[year] += co2e
```

```python
print(co2_e)
```

```python
co2e_df = pd.DataFrame.from_dict(co2_e, orient='index')
co2e_df
```

```python
co2e_df.plot()
```

# California Non-Renewable Energy Tables

Importing data from 2014-2019.

```python
energy_df = pd.read_csv("./data/non-renewable-energy/CaliforniaEnergyTableAggregate2014-2019.csv")
```

```python
energy_df.shape
```

```python
energy_df.describe(include='all')
```

Display subset of table only showing renewable vs non-renewable energy production per year.

```python
energy_df.columns
```

```python
energy_sub = pd.DataFrame(energy_df, copy=True, columns=['Fuel Type', 'Total California Energy Mix (GWh)', 'Year'])
energy_sub.shape
```

```python
energy_sub.dropna(how="all", inplace=True)
energy_sub.shape
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
energy_sub
```

```python
energy_sub.describe(include="all")
```

```python
energy_sub['Fuel Type'].unique()
```

```python
diff_names = ['Non-Renewables and Unspecified Totals', 'Renewables Totals']
```

```python
energy_min = []
for key, row in energy_sub.iterrows():
    if row['Fuel Type'] in diff_names:
        energy_min.append(row)
```
